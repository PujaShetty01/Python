# Python Basic Questions & answers

# 1. What is Python, and why is it popular?
 Python is a high-level, interpreted, general-purpose programming language. It is popular because of its simple syntax, readability, large community support, and wide usage in fields like data science, web development, and automation.

# 2. What is an interpreter in Python?
An interpreter executes Python code line by line and directly translates it into machine instructions.

# 3. What are pre-defined keywords in Python?
Keywords are reserved words in Python with special meanings (e.g., if, for, class, def). They cannot be used as identifiers.

# 4. Can keywords be used as variable names?
No, keywords cannot be used as variable names.

# 5. What is mutability in Python?
Mutability means whether an object can be modified after creation.

Mutable → Lists, dictionaries, sets.

Immutable → Tuples, strings, integers.

# 6. Why are lists mutable, but tuples are immutable?
Lists allow modification (add, remove, update elements). Tuples are fixed in nature, created for data safety and faster performance, hence immutable.

# 7. Difference between == and is operators in Python?

== → Checks if values are equal.

is → Checks if two variables point to the same object in memory.

# 8. What are logical operators in Python?

and → True if both conditions are True.

or → True if at least one condition is True.

not → Negates the condition.

# 9. What is type casting in Python?
Type casting means converting one data type into another (e.g., string → integer).

# 11. Purpose of conditional statements in Python?
They allow decision-making by executing code only when a condition is true.

# 12. How does the elif statement work?
It checks multiple conditions one after another. As soon as one condition is true, its block executes and the rest are skipped.

# 13. Difference between for and while loops?

for → Used when number of iterations is known.

while → Used when iterations depend on a condition.

# 14. Scenario where a while loop is more suitable?
When the number of iterations is not fixed, e.g., repeatedly asking for a password until correct.

# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify(a_list, a_num):
    a_list.append(100)   # Mutable (list will change outside function)
    a_num += 10          # Immutable (int will not change outside function)

lst = [1, 2]
num = 5

print("Before:", lst, num)
modify(lst, num)
print("After:", lst, num)
```

# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print("Sum of even numbers between 1 and 50:", total)
```

# 14. Write a program to reverse a string using a while loop.

```python
s = input("Enter a string: ")
rev = ""
i = len(s) - 1

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed string:", rev)
```

# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
n = int(input("Enter a number: "))
fact = 1

while n > 0:
    fact *= n
    n -= 1

print("Factorial:", fact)
```
